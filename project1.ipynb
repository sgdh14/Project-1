```python
def convert_text_to_numeric(df_data,column):
    df_data[column] = pd.to_numeric(df_data[column],errors='coerce')

for column in text_columns:
    convert_text_to_numeric(df_data, column)
    print(f'{column}:{df_data[column].dtype}')
```

```python
def mean_value_imputation(df_data,column):
    mean_value = df_data[column].mean()
    df_data[column].fillna(value=mean_value,inplace=True)


def mode_value_imputation(df_data,column):
    mode = df_data[column].mode()[0]
    df_data[column] = df_data[column].fillna(mode)
```

```python
num_cols=[col for col in df_data.columns if df_data[col].dtype != 'object']
for col_name in num_cols:
    mean_value_imputation(df_data,col_name)
```

```python
cat_cols=[col for col in df_data.columns if df_data[col].dtype == 'object']
for col_name in cat_cols:
    mode_value_imputation(df_data,col_name)
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier # Make sure to import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score, recall_score, f1_score
from sklearn.pipeline import Pipeline # Import Pipeline for easier workflow

# ... (Your existing code for loading and splitting data) ...

# Create a pipeline with imputation and your models
model = []
model.append(('GaussianNB', Pipeline([('imputer', SimpleImputer()), ('gnb', GaussianNB())])))
model.append(('KNeighborsClassifier', Pipeline([('imputer', SimpleImputer()), ('knn', KNeighborsClassifier(n_neighbors=8))])))
model.append(('DecisionTreeClassifier', Pipeline([('imputer', SimpleImputer()), ('dtc', DecisionTreeClassifier())])))
model.append(('SVC', Pipeline([('imputer', SimpleImputer()), ('svc', SVC(kernel='linear'))])))
model.append(('RandomForestClassifier', Pipeline([('imputer', SimpleImputer()), ('rfc', RandomForestClassifier())])))

# Iterate through the models and fit them to the training data
for name, model in model:
    print(f"Model Name : {name}")
    model.fit(x_train, y_train)  # The imputer is automatically applied
    y_pred = model.predict(x_test)
    print(f"Accuracy Score : {accuracy_score(y_test, y_pred)}")
    # print('\n')
    print(f"Precision Score : {precision_score(y_test, y_pred)}")
    # print('\n')
    print(f"Recall Score : {recall_score(y_test, y_pred)}")
    # print('\n')
    print(f"F1 Score : {f1_score(y_test, y_pred)}")
    # print('\n')
    print(f"Confusion Matrix : {confusion_matrix(y_test, y_pred)}")
    print('\n')
```
